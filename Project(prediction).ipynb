```python
# Preprocessing
```

```python
df
```

```python
# Define a function to categorize the time of day
def categorize_time(hour):
    if 5 <= hour < 8:
        return 'Early Morning'
    elif 8 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 16:
        return 'Afternoon'
    elif 16 <= hour < 20:
        return 'Evening'
    else:
        return 'Night'

# Apply the function to Arrival_hours and Dep_hours columns
df['Arrival_Time_Period'] = df['Arrival_hours'].apply(categorize_time)
df['Dep_Time_Period'] = df['Dep_hours'].apply(categorize_time)

# Drop the original hours columns if you no longer need them
df.drop(['Arrival_hours', 'Dep_hours', 'Dep_min',	'Arrival_min',	'Duration_hours',	'Duration_min'	], axis=1, inplace=True)

df
```

```python
df.drop(['Dep_min', 'Arrival_min','Duration_hours',	'Duration_min'], axis=1, inplace=True)
df
```

```python
df.Total_Stops = pd.factorize(df.Total_Stops)[0]
```

```python
df=df.join(pd.get_dummies(df.Airline, prefix='Airline')).drop('Airline', axis=1)
```

```python
df=df.join(pd.get_dummies(df.Source, prefix='From')).drop('Source', axis=1)
df=df.join(pd.get_dummies(df.Destination, prefix='To')).drop('Destination', axis=1)
df=df.join(pd.get_dummies(df.Arrival_Time_Period, prefix='Arrival')).drop('Arrival_Time_Period', axis=1)
df=df.join(pd.get_dummies(df.Dep_Time_Period, prefix='Departure')).drop('Dep_Time_Period', axis=1)
```

```python
df = df.astype(int)
```

```python
df
```

```python
# Training
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
x,y = df.drop('Price',axis=1), df.Price
```

```python
x_train, x_test, y_train, y_test = train_test_split (x,y,test_size=0.2)
```

```python
reg = RandomForestRegressor(n_jobs=-1)
reg.fit(x_train, y_train)
```

```python
reg.score(x_test, y_test)
```

```python
import math
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_pred = reg.predict(x_test)

print('R2:' , r2_score(y_test,y_pred))
print('MAE:' , mean_absolute_error(y_test,y_pred))
print('MSE:' , mean_squared_error(y_test,y_pred))
print('RMSE:' , math.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test,y_pred)
plt.xlabel('Actual Flight Price')
plt.ylabel('Predicted Flight Price')
plt.title('Prediction VS Actual Price')
```
